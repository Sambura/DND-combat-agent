```python
import numpy as np
import random
import sys

if '..' not in sys.path: sys.path.append('..')
from src.utils.common import *
from src.dnd.actions import *
from src.dnd.units import *
from src.dnd.game_utils import *
from src.dnd.game_board import DnDBoard
```

```python
game = DnDBoard()
ally = GenericSoldier("Ally soldier")
enemy = GenericSoldier("Enemy soldier")
game.place_unit(ally, (2, 2), 0)
game.place_unit(enemy, (7, 7), 1)
game.initialize_game()
game.turn_order = [0, 1]

color_map = {
    ally: "Green",
    enemy: "Red",
}
```

```python
print_game(game, color_map)
```

```python
unit, player_id = game.get_current_unit()
new_coords = (3, 3)
game.take_turn(new_coords, None)

print_game(game, color_map)
```

```python
unit, player_id = game.get_current_unit()
new_coords = (5, 6)
game.take_turn(new_coords, None)

print_game(game, color_map)
```

```python
unit, player_id = game.get_current_unit()
new_coords = (4, 5)
action_instance = ActionInstance(None and unit.actions[0], source_unit=unit, target_unit=enemy)
game.take_turn(new_coords, action_instance)

print_game(game, color_map)
```

```python
unit, player_id = game.get_current_unit()
new_coords = (5, 5)
action_instance = ActionInstance(unit.actions[0], source_unit=unit, target_unit=ally)
game.take_turn(new_coords, action_instance)

print_game(game, color_map)
```

```python
for i in range(8):
    unit, player_id = game.get_current_unit()
    new_coords = (4, 5)
    action_instance = ActionInstance(unit.actions[0], source_unit=unit, target_unit=enemy)
    game.take_turn(new_coords, action_instance)

    unit, player_id = game.get_current_unit()
    new_coords = (5, 5)
    action_instance = ActionInstance(unit.actions[0], source_unit=unit, target_unit=ally)
    game.take_turn(new_coords, action_instance)

    print_game(game, color_map)
```
